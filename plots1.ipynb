```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
data= sorted([random.randint(135, 155) for _ in range(40)])
print(data)

# Create a histogram
plt.hist(data, color='skyblue', edgecolor='black')

# Add labels and title
plt.xticks(rotation=45)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Display the histogram
plt.show()
```

```python
plt.plot(data)
plt.show()
```

```python
expense = {
    "electricity":5000,
    "food":10000,
    "entertainment":1500,
    "travel":500,
    "others":2000
}

plt.pie(expense.values(), labels=expense.keys(), autopct="%0.2f%%")
# plt.legend(expense.keys())
plt.show()
```

```python
plt.bar(expense.keys(), expense.values())
plt.show()
```

```python
h = sorted([random.randint(135, 155) for _ in range(40)])
w = sorted([random.uniform(40, 50) for _ in range(40)])



plt.scatter(h, w)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```
